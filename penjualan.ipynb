```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("penjualan.csv")

data.head()
data.info()
data.describe()
```

```python
data['Tanggal'] = pd.to_datetime(data['Tanggal'])
```

Tugasnya adalah:

üì• Baca file penjualan.csv ke pandas
‚ûó Hitung kolom Total_Penjualan = Jumlah * Harga_Satuan
üìä Group by Produk ‚Üí total jumlah terjual & pendapatan
üìà Visualisasi bar chart penjualan
üß† Insight akhir
üíæ Simpan hasil ke CSV
‚úçÔ∏è Perbaiki README.md dan upload ke GitHub

```python
data['Total_Penjualan']= data['Jumlah']*data['Harga_Satuan']

data.head()
```

```python
# Total penjualan tiap produk
penjualan_per_produk = data.groupby('Produk')['Total_Penjualan'].sum().sort_values(ascending=False)
print(penjualan_per_produk)
```

```python
jumlah_terjual = data.groupby('Produk')['Jumlah'].sum().sort_values(ascending=False)
print(jumlah_terjual)
```

```python
penjualan_per_produk.plot(kind='bar', color='skyblue', figsize=(10, 5))
plt.title('Total Pendapatan per Produk')
plt.ylabel('Total (Rp)')
plt.xlabel('Produk')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
jumlah_terjual.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), startangle=90)
plt.title('Distribusi Jumlah Barang Terjual per Produk')
plt.ylabel('')
plt.show()
```

# üìå Insight

- Produk dengan total penjualan terbesar adalah **Baju**. Selain paling menguntungkan, produk ini juga menjadi barang yang **paling laris**.
- Produk dengan harga mahal seperti **Jam Tangan** ternyata memiliki volume penjualan rendah, termasuk dalam **tiga produk paling sedikit terjual**.
- Berdasarkan pola ini, disarankan agar toko lebih memfokuskan strategi penjualan pada produk-produk yang **tinggi permintaan namun tetap menguntungkan**, yaitu: **Baju**, **Topi**, dan **Celana**.
